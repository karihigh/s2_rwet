# Generating Décimas
## A poetic style composed of 10 eight-syllable-lines stanzas
### With a rhyming scheme of ABBAACCDDC

For this final assignment I decided to make a program that could generate this style of composing a poem (or song). Since this form is spread mostly among Latin America, I used as input text the introduction of a historical essay by Eduardo Galeano. 

````python
import markovify
import pronouncing as pr
from collections import Counter
import re
import tracery
import random
import spacy
````

````python
nlp = spacy.load('en_core_web_md')
````

````python
text = open("veins.txt").read()
````

````python
text.replace("\n", " ").replace(",", " ").replace(".", " ").replace("\\", " ").replace("-", " ").replace("—", " ")
````

````python
doc = nlp(text)
````

````python
words = [item.text for item in doc]
````

### List of each word of the text with its pronunciation

````python
phones_list = [(pr.phones_for_word(w)[0], w) for w in words if len(pr.phones_for_word(w))>0]
````

````python
phones_list
````

````python
print(phones_list[1])
````

### Same list but now with only the rhyming part of the pronunciation

````python
rh_list = [(pr.rhyming_part(r[0]),r[1]) for r in phones_list]
````

````python
rh_list
````

````python
print(rh_list[1])
````

### Making a dictionary 

````python
rh_dict = {}
for rhy in rh_list:
    rhy_part = rhy[0] # rhyming part
    # check if it is already in the dict
    if(rhy_part not in rh_dict.keys()):
        rh_dict[rhy_part] = [] #create the list if it doesn´t exist
    rh_dict[rhy_part].append(rhy[1].lower()) # add the word to the list

# now, purge duplicates
for rhy in rh_dict.keys():
    rh_dict[rhy] = list(set(rh_dict[rhy]))
````

````python
rh_dict
````

### Selecting only the rhyming part that repeats at least three times

````python
common_rhymes = {}
for rhy in rh_dict.keys():
    if len(rh_dict[rhy])>3 :
        common_rhymes[rhy] = rh_dict[rhy]
````

````python
rh_list = []
[rh_list.extend(list_of_vals) for list_of_vals in common_rhymes.values()]
print(rh_list)
````

### Training the markov model

````python
def ngrams_for_sequence(n, seq):
    return[tuple(seq[i:i+n]) for i in range(len(seq)-n+1)]
````

````python
words = [w.text for w in doc if w.pos_ not in ["PUNCT", "SPACE", "x"]]
````

````python
chunks = ngrams_for_sequence(6, words)
````

````python
chunks
````

````python
chunks_as_strings = [' '.join(item) for item in chunks]
````

````python
chunks_as_strings
````

````python
lines = chunks_as_strings #text.split(".")
````

````python
reversed_by_line = "\n".join([' '.join(reversed(line.split())) for line in lines])
````

````python
gen = markovify.NewlineText(reversed_by_line, state_size=1)
````

### Function to generate lines using the words from the generated list of most rhymed words

````python
def syllable_count(s):
    words = [word.strip(',.?:').lower() for word in s.split()]
    print(words)
    return sum([pr.syllable_count(pr.phones_for_word(item)[0]) for item in words])
````

````python
def gen_syllables(gen, init_state, n=10, tries=3000):
    for i in range(tries):
        s = gen.make_sentence(test_output=False, init_state=init_state)
        if syllable_count(s) == n:
            return s
    return None
````

````python
def gen_dec(i, tries):
    init_word = rh_list[i]
    result = ""
    for x in range(tries):
        try:
            backwards = gen_syllables(gen, (init_word,))
        except:
            #print("!!!\t crashed on " + init_word)
            continue
        back_split = backwards.split(" ")
        forwards_split = reversed(back_split)
        result = " ".join(forwards_split)
        break
    return result
````

````python
dec_sents = []
[dec_sents.append(gen_dec(i, 300)) for i in range(len(rh_list))]
````

````python
dec_sents2 = []
[dec_sents2.append(gen_dec(i, 300)) for i in range(len(rh_list))]
````
